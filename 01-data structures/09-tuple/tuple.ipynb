Store the information of three students (first name, last name, age, and score) in an appropriate data structure.

```python
students = [
    ("Emma", "Miller", 18, 78),
    ("Liam", "Garcia", 19, 90),
    ("Ethan", "Davis", 18, 99)
]
name, family, age, score = students[0]
for name, family, age, score in students:
    print(f"My name is: {name}{family} and I'm {age} years old. My score in university is: {score}")
f"My name is: {name}{family} and I'm {age} years old. My score in university is: {score}"
```
