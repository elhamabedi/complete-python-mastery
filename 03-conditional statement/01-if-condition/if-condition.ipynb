### Exercise – Workout Points Game

Write a program that:

- Sets `action = "exercise"`.
- Saves the score in a file.
- If `action` is `"run"` or `"exercise"`, increases the score and prints an appropriate message.
- If `action` is `"skip"`, the score does not change.
- For any other value, prints an error message.
- At the end, prints a success or failure message based on the final score.

(Choose your own message texts and file name.)

```python
try:
    with open("score.txt", "r") as f:
        score = int(f.read())
except:
    score = 0

action = "exercise"

if action == "run":
    print("you ran 1 km!")
    score +=1
elif action == "exercise":
    print("nice workout!")
    score +=1
else:
    print("you skipped training!")

with open("score.txt", "w") as f:
    f.write(str(score))

if score > 0:
    print(f"well done! total points: {score}")
else:
    print("no training yet. start moving!")
```
