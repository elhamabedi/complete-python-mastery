### Exercise – Power Function with Default Parameter

Write a function named `power` that takes two parameters:
- `n`: the base number
- `t`: the exponent (default value = `4`)

The function should return `n` raised to the power of `t`.

Then, take a number from the user as input, call the function with that number (using the default exponent), and print the result.

````python
def power(n, t=4):
    result = 1
    for i in range(t):
        result *= n
    return result

user_number = int(input("Enter your number: "))
print(power(user_number))
````

### Exercise – Count Even Numbers with Helper Function

Write a function that takes a list of numbers and returns the count of even numbers in the list.  
Inside this function, define a **helper function** (inner function) that checks whether a given number is even.

Then, call the function with any list of your choice and print the result.

````python
def number_of_even(old_list):
    def is_even(n):
        return (n % 2 ==0)
    count = 0
    for n in old_list:
        if is_even(n):
            count +=1
    return count


count = number_of_even( [34, 78, 999, 2, 3, 1000, 0])
print(count)
````

### Exercise – Find the Largest Number

Write a function that takes a list of numbers and returns the largest number in the list.

````python
def max_number(list):
    maximum = list[0]
    for i in list:
        if i > maximum:
            maximum = i

    return maximum
        
max_number([5,8,2,3,90,76])
````

Write a function that takes a list of numbers and returns their average.

```
student_marks = [20, 15, 10, 11.75, 14.5, 19.25, 17, 17, 17, 12]
```

````python
def average(student_marks):
    total = 0               
    count = len(student_marks)
    for mark in student_marks:
        total = total + mark
    avg = total / count
    return avg

result = average([20, 15, 10, 11.75, 14.5, 19.25, 17, 17, 17, 12])
print(result)
````
