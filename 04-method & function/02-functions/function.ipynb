### Exercise – Power Function with Default Parameter

Write a function named `power` that takes two parameters:
- `n`: the base number
- `t`: the exponent (default value = `4`)

The function should return `n` raised to the power of `t`.

Then, take a number from the user as input, call the function with that number (using the default exponent), and print the result.

```python
def power(n, t=4):
    result = 1
    for i in range(t):
        result *= n
    return result

user_number = int(input("Enter your number: "))
print(power(user_number))
```
